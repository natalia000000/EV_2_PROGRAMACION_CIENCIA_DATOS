```python
#import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report,
    roc_auc_score,
    roc_curve
)
```

**`03_model_evaluation.ipynb`** se enfoca en evaluar y comparar el rendimiento de los modelos supervisados entrenados previamente, como **Logistic Regression, Decision Tree y Random Forest**. En este notebook se calculan métricas clave como **Accuracy, Precision, Recall, F1-Score y ROC-AUC**, además de generar visualizaciones como la **matriz de confusión** y la **curva ROC**. Su objetivo principal es interpretar los resultados obtenidos, identificar fortalezas y debilidades de cada modelo, y determinar cuál ofrece el mejor desempeño para la predicción de videojuegos exitosos (`Exitoso`).


```python
#Predicciones
y_pred_lr = lr.predict(X_test_scaled)
y_pred_dt = dt.predict(X_test)
y_pred_rf = rf.predict(X_test)
```

```python
#metricas
results = pd.DataFrame({
    "Model": ["Logistic Regression", "Decision Tree", "Random Forest"],
    "Accuracy": [
        accuracy_score(y_test, y_pred_lr),
        accuracy_score(y_test, y_pred_dt),
        accuracy_score(y_test, y_pred_rf)
    ],
    "Precision": [
        precision_score(y_test, y_pred_lr),
        precision_score(y_test, y_pred_dt),
        precision_score(y_test, y_pred_rf)
    ],
    "Recall": [
        recall_score(y_test, y_pred_lr),
        recall_score(y_test, y_pred_dt),
        recall_score(y_test, y_pred_rf)
    ],
    "F1": [
        f1_score(y_test, y_pred_lr),
        f1_score(y_test, y_pred_dt),
        f1_score(y_test, y_pred_rf)
    ]
})

results
```

```python
#Matriz de confusion
cm = confusion_matrix(y_test, y_pred_rf)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d')
plt.title("Confusion Matrix - Random Forest")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
#Reporte de clasificacion
print(classification_report(y_test, y_pred_rf))
```

```python
#ROC Curve
y_prob = rf.predict_proba(X_test)[:,1]

fpr, tpr, _ = roc_curve(y_test, y_prob)

plt.plot(fpr, tpr)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.show()
```
