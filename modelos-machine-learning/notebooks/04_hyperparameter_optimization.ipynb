Este notebook carga el mejor modelo previamente entrenado (`best_model.pkl`), aplica **GridSearchCV** para probar distintas combinaciones de hiperparámetros, selecciona la mejor configuración según el **F1-score**, evalúa su rendimiento con los datos de prueba y finalmente guarda el modelo optimizado como `optimized_model.pkl` para usarlo posteriormente.


```python
import pandas as pd
import joblib

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)


# CARGAR DATOS

X_train = pd.read_csv("../data/05_model_input/X_train.csv")
X_test = pd.read_csv("../data/05_model_input/X_test.csv")
y_train = pd.read_csv("../data/05_model_input/y_train.csv").squeeze()
y_test = pd.read_csv("../data/05_model_input/y_test.csv").squeeze()


# CARGAR MODELO BASE

model = joblib.load("../data/06_models/best_model.pkl")

print("Modelo cargado correctamente")
print(model)
```

```python
#RandomForest
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [5, 10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
#RandomForest
param_grid = {
    'max_depth': [3, 5, 10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}
```

```python
#DecisionTree
param_grid = {
    'max_depth': [3, 5, 10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}
```

```python
#LogisticRegression
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'lbfgs'],
    'penalty': ['l2']
}
```

```python
#GridSearchCV
grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    cv=5,
    scoring='f1',
    n_jobs=-1,
    verbose=2
)

grid_search.fit(X_train, y_train)
```

```python
#Mejor modelo
best_model = grid_search.best_estimator_

print("Mejores parámetros:")
print(grid_search.best_params_)
```

```python
#Evaluación del modelo optimizado
y_pred = best_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1 Score :", f1)
```

```python
#Modelo optimizado guardado
joblib.dump(
    best_model,
    "../data/06_models/optimized_model.pkl"
)

print("Modelo optimizado guardado")
```
