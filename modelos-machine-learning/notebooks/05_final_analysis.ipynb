Para `05_final_analysis.ipynb`, normalmente se hace el **análisis final del modelo optimizado**: cargar el modelo, evaluar métricas finales, visualizar resultados y sacar conclusiones.

### Pequeño resumen:

Este notebook carga el modelo optimizado (`optimized_model.pkl`), realiza las predicciones finales sobre los datos de prueba, evalúa métricas como **Accuracy, Precision, Recall, F1-score y ROC-AUC**, genera visualizaciones como la **matriz de confusión** y la **curva ROC**, y finalmente presenta una conclusión sobre el rendimiento general del modelo y si está listo para ser usado o necesita mejoras adicionales.


```python
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report,
    roc_curve
)

# CARGAR DATOS
X_test = pd.read_csv("../data/05_model_input/X_test.csv")
y_test = pd.read_csv("../data/05_model_input/y_test.csv")

# Convertir y_test a vector
y_test = y_test.squeeze()

print("Datos cargados correctamente")
print("X_test:", X_test.shape)
print("y_test:", y_test.shape)
```

```python
# CARGAR MODELO OPTIMIZADO
model = joblib.load("../data/06_models/optimized_model.pkl")

print("Modelo optimizado cargado correctamente")
print(model)
```

```python
# PREDICCIONES
y_pred = model.predict(X_test)

# Probabilidades (para ROC-AUC)
y_prob = model.predict_proba(X_test)[:, 1]
```

```python
# MÉTRICAS FINALES
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_prob)

print("========== RESULTADOS FINALES ==========")
print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1 Score :", f1)
print("ROC-AUC  :", roc_auc)
```

```python
# MATRIZ DE CONFUSIÓN
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(
    cm,
    annot=True,
    fmt="d"
)

plt.title("Confusion Matrix - Final Model")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# CLASSIFICATION REPORT
print("========== CLASSIFICATION REPORT ==========")
print(classification_report(y_test, y_pred))
```

```python
# CURVA ROC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.plot([0,1], [0,1], '--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Final Model")
plt.legend()
plt.show()
```

```python
# COMPARACIÓN REAL VS PREDICHO
results = pd.DataFrame({
    "Actual": y_test,
    "Predicted": y_pred
})

print(results.head(20))
```

```python
# CONCLUSIÓN FINAL
print("========== ANÁLISIS FINAL ==========")

if f1 >= 0.80:
    print("El modelo presenta un excelente rendimiento y puede considerarse robusto.")
elif f1 >= 0.70:
    print("El modelo tiene buen rendimiento, aunque podría optimizarse.")
elif f1 >= 0.60:
    print("El modelo es aceptable, pero requiere mejoras.")
else:
    print("El modelo necesita una optimización importante.")
```
